```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('A회사_보습제_매출데이터_v1.csv', encoding='cp949')
print(df)
```

```python
for column in df.iloc[:, 1:]:
    df[column] = df[column].replace('[\$,]', '', regex=True).astype(float)

date_range = df.columns[1:]
total_sales_data = df.iloc[0, 1:]

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(date_range, total_sales_data, marker='o')

ax.set_title('Total Sales Trend for 6 Months')
ax.set_xlabel('Month')
ax.set_ylabel('Sales (USD)')
ax.set_xticklabels(date_range, rotation=45)
ax.grid(True)

plt.show()
```

```python
monthly_ad_costs = df.iloc[2, 1:]

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(date_range, monthly_ad_costs, marker='o', color='red')

ax.set_title('Monthly Ad Costs')
ax.set_xlabel('Month')
ax.set_ylabel('Cost (USD)')
ax.set_xticklabels(date_range, rotation=45)
ax.grid(True)

plt.show()
```

```python
total_sales = df.loc[0, date_range]
goal_sales = df.loc[1, date_range]
monthly_costs = df.loc[2, date_range]
sns_ad_costs = df.loc[3, date_range]

fig1, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(date_range, total_sales, marker='o', label='Total Sales')
ax1.plot(date_range, goal_sales, marker='o', label='Goal Sales')
ax1.plot(date_range, monthly_costs, marker='o', label='Ad Costs')
ax1.plot(date_range, sns_ad_costs, marker='o', label='SNS Ad Costs')

ax1.set_title('Sales, Goal Sales, Ad Costs, and SNS Ad Costs')
ax1.set_xlabel('Month')
ax1.set_ylabel('Amount (USD)')
ax1.set_xticklabels(date_range, rotation=45)
ax1.grid(True)
ax1.legend()

plt.show()


fig2, (ax2, ax3) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

ax2.plot(date_range, total_sales, marker='o', label='Total Sales')
ax2.plot(date_range, goal_sales, marker='o', label='Goal Sales')
ax2.set_ylim(5000000, 7000000) 
ax2.grid(True)
ax2.legend()

ax3.plot(date_range, monthly_costs, marker='o', label='Ad Costs')
ax3.plot(date_range, sns_ad_costs, marker='o', label='SNS Ad Costs')
ax3.set_ylim(0, 1250000) 
ax3.grid(True)
ax3.legend()

fig2.suptitle('Sales, Goal Sales, Ad Costs, and SNS Ad Costs')
plt.xticks(rotation=45)
plt.xlabel('Month')

plt.show()
```

```python
price = df.loc[4, date_range]

plt.figure(figsize=(10, 6))
plt.plot(date_range, price, marker='o', linestyle='-', color='green')
plt.title('Unit Price per Ounce')
plt.xlabel('Month')
plt.ylabel('Unit Price (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
